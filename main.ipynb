```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import accuracy_score
# import datawig
```

```python
df2 = pd.read_json('./dataSet_Culture_06102023-POINT.json')
```

```python
df2 = df2.drop(columns=['type_culture_name', 'soil_id'])
```

```python
print(df2.isnull().sum())
print(df2.shape)
# df2['type_culture_name'] = df2['type_culture_name'].replace(np.NaN, 'nodataculturetype')
```

```python
df2['Analysis Date'].info
```

```python
df2['Analysis Date'] = pd.to_datetime(df2['Analysis Date'])
```

```python
df2['x_coordinate'] = df2['polygon'].apply(lambda coord: coord['x'])
df2['y_coordinate'] = df2['polygon'].apply(lambda coord: coord['y'])

# If you want to drop the original 'polygon' column
df2 = df2.drop(columns=['polygon'])
```

```python
df2['combined'] = df2['x_coordinate'].astype(str) + '_' + df2['y_coordinate'].astype(str) + '_' + df2['year contour'].astype(str)

# Assign unique ID based on the grouped column
df2['field_id'] = df2.groupby('combined').ngroup() + 1

# Drop the combined column
df2 = df2.drop(columns=['combined'])

df2
```

```python
# Apply label encoding
for column in df2.columns:
    le = LabelEncoder()
    df2[column + '_encoded'] = le.fit_transform(df2[column])

print(df2)
```
