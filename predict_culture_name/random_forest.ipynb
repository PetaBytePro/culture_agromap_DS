Decision Tree

```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_json('../data/dataSet_Culture_06102023-POINT.json')
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df['x'] = df['polygon'].apply(lambda row: row['x'])
df['y'] = df['polygon'].apply(lambda row: row['y'])

df['date'] = pd.to_datetime(df['Analysis Date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df = df.drop(['polygon', 'Analysis Date', 'year contour', 'date'], axis='columns')
df.head()
```

```python
inputs = df.drop('culture_name', axis='columns')
target = df['culture_name']
```

```python
inputs.head()
```

```python
le_indextype = LabelEncoder()
le_district_name = LabelEncoder()
le_soil_name = LabelEncoder()
le_type_culture_name = LabelEncoder()
```

```python
inputs['index_type_n'] = le_indextype.fit_transform(inputs['indextype'])
inputs['district_name_n'] = le_district_name.fit_transform(inputs['district_name'])
inputs['soil_name_n'] = le_soil_name.fit_transform(inputs['soil_name'])
inputs['type_culture_name_n'] = le_type_culture_name.fit_transform(inputs['type_culture_name'])
```

```python
inputs.head()
```

```python
inputs_n = inputs.drop(['indextype', 'district_name', 'soil_name', 'type_culture_name'], axis='columns')
inputs_n.head()
```

Train

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

Predict

```python
model.score(inputs_n, target)
```
